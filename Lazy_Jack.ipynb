```python
import os
from unify import Unify
import pandas as pd
import textstat

os.environ["UNIFY_KEY"] = "zgZRC0qqOC89cjdiDq403mtdgp-n8ULnQjK6JMiS2YE="

def generate_response(model_name, prompt, temperature=0.7):
    unify = Unify(model_name)
    response = unify.generate(
        prompt,
        temperature=temperature
    )
    return response

models = [
    "claude-3-haiku@anthropic",
    "claude-3-opus@anthropic",
    "claude-3.5-sonnet@anthropic",
    "gemini-1.5-flash@vertex-ai",
    "gemini-1.5-pro@vertex-ai",
    "gemma-2-9b-it@fireworks-ai",
    "gemma-2b-it@together-ai",
    "gpt-3.5-turbo@openai",
    "gpt-4@openai",
    "gpt-4-turbo@openai",
    "gpt-4o@openai",
    "llama-3-70b-chat@fireworks-ai",
    "llama-3-8b-chat@fireworks-ai",
    "mistral-7b-instruct-v0.3@together-ai",
    "mistral-large@aws-bedrock",
    "mistral-small@mistral-ai",
    "mixtral-8x22b-instruct-v0.1@deepinfra",
    "mixtral-8x7b-instruct-v0.1@aws-bedrock",
    "qwen-2-72b-instruct@deepinfra"
]

prompt = "You are a fairy tale author. Write a fairy tale about LAZY JACK"


original_text = """
Once upon a time there was a boy whose name was Jack, and he lived with his mother on a common. They were very poor, and the old woman got her living by spinning, but Jack was so lazy that he would do nothing but bask in the sun in the hot weather, and sit by the corner of the hearth in the winter-time. So they called him Lazy Jack. His mother could not get him to do anything for her, and at last told him, one Monday, that if he did not begin to work for his porridge she would turn him out to get his living as he could.
This roused Jack, and he went out and hired himself for the next day to a neighbouring farmer for a penny; but as he was coming home, never having had any money before, he lost it in passing over a brook. “You stupid boy,” said his mother, “you should have put it in your pocket.” “I'll do so another time,” replied Jack.
On Wednesday, Jack went out again and hired himself to a cow-keeper, who gave him a jar of milk for his day's work. Jack took the jar and put it into the large pocket of his jacket, spilling it all, long before he got home. “Dear me!” said the old woman; “you should have carried it on your head.” “I'll do so another time,” said Jack.
So on Thursday, Jack hired himself again to a farmer, who agreed to give him a cream cheese for his services. In the evening Jack took the cheese, and went home with it on his head. By the time he got home the cheese was all spoilt, part of it being lost, and part matted with his hair. “You stupid lout,” said his mother, “you should have carried it very carefully in your hands.” “I'll do so another time,” replied Jack.
On Friday, Lazy Jack again went out, and hired himself to a baker, who would give him nothing for his work but a large tom-cat. Jack took the cat, and began carrying it very carefully in his hands, but in a short time pussy scratched him so much that he was compelled to let it go. When he got home, his mother said to him, “You silly fellow, you should have tied it with a string, and dragged it along after you.” “I'll do so another time,” said Jack.
So on Saturday, Jack hired himself to a butcher, who rewarded him by the handsome present of a shoulder of mutton. Jack took the mutton, tied it to a string, and trailed it along after him in the dirt, so that by the time he had got home the meat was completely spoilt. His mother was this time quite out of patience with him, for the next day was Sunday, and she was obliged to make do with cabbage for her dinner. “You ninney-hammer,” said she to her son; “you should have carried it on your shoulder.” “I'll do so another time,” replied Jack.
On the next Monday, Lazy Jack went once more, and hired himself to a cattle-keeper, who gave him a donkey for his trouble. Jack found it hard to hoist the donkey on his shoulders, but at last he did it, and began walking slowly home with his prize. Now it happened that in the course of his journey there lived a rich man with his only daughter, a beautiful girl, but deaf and dumb. Now she had never laughed in her life, and the doctors said she would never speak till somebody made her laugh. This young lady happened to be looking out of the window when Jack was passing with the donkey on his shoulders, with the legs sticking up in the air, and the sight was so comical and strange that she burst out into a great fit of laughter, and immediately recovered her speech and hearing. Her father was overjoyed, and fulfilled his promise by marrying her to Lazy Jack, who was thus made a rich gentleman. They lived in a large house, and Jack's mother lived with them in great happiness until she died.






"""

results_df = pd.DataFrame(columns=[
    "model", "metric", "value"
])

def calculate_metrics(text, model_name):
    metrics = {
        "Syllable Count": textstat.syllable_count(text),
        "Lexicon Count": textstat.lexicon_count(text, removepunct=True),
        "Sentence Count": textstat.sentence_count(text),
        "Flesch Reading Ease": textstat.flesch_reading_ease(text),
        "Flesch-Kincaid Grade Level": textstat.flesch_kincaid_grade(text),
        "Gunning Fog": textstat.gunning_fog(text),
        "SMOG Index": textstat.smog_index(text),
        "Automated Readability Index": textstat.automated_readability_index(text),
        "Coleman-Liau Index": textstat.coleman_liau_index(text),
        "Linsear Write Formula": textstat.linsear_write_formula(text),
        "Dale-Chall Readability Score": textstat.dale_chall_readability_score(text),
        "Text Standard": textstat.text_standard(text, float_output=False),
        "Reading Time (seconds)": textstat.reading_time(text)
    }
    return metrics

original_metrics = calculate_metrics(original_text, "original_text")
for metric, value in original_metrics.items():
    results_df = pd.concat([results_df, pd.DataFrame([{
        "model": "original_text",
        "metric": metric,
        "value": value
    }])], ignore_index=True)

for model in models:
    print(f"\nResponse from {model}:")
    try:
        response = generate_response(model, prompt)
        print(response)

        model_metrics = calculate_metrics(response, model)
        for metric, value in model_metrics.items():
            results_df = pd.concat([results_df, pd.DataFrame([{
                "model": model,
                "metric": metric,
                "value": value
            }])], ignore_index=True)

        print("\nReadability Metrics:")
        for metric, value in model_metrics.items():
            print(f"{metric}: {value}")

    except Exception as e:
        print(f"Error generating response from {model}: {e}")

results_df.to_csv("LAZY JACK_readability_score.csv", index=False)
results_df.to_excel("LAZY JACK_readability_score.xlsx", index=False)


print("Responses and readability metrics saved to LAZY JACK_readability_score.csv, LAZY JACK_readability_score.xlsx")
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("/content/LAZY JACK_readability_score.csv")

metric_directions = {
    "Syllable Count": "lower",
    "Lexicon Count": "higher",
    "Sentence Count": "higher",
    "Flesch Reading Ease": "higher",
    "Flesch-Kincaid Grade Level": "lower",
    "Gunning Fog": "lower",
    "SMOG Index": "lower",
    "Automated Readability Index": "lower",
    "Coleman-Liau Index": "lower",
    "Linsear Write Formula": "lower",
    "Dale-Chall Readability Score": "lower",
    "Text Standard": "lower",
    "Reading Time (seconds)": "lower"
}

pivot_df = df.pivot(index='model', columns='metric', values='value')
if 'Text Standard' in pivot_df.columns:
    pivot_df = pivot_df.drop("Text Standard", axis=1)

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

for metric, direction in metric_directions.items():
    if direction == "lower" and metric in normalized_df.columns:
        normalized_df[metric] = 1 - normalized_df[metric]

normalized_df['Composite Score'] = normalized_df.mean(axis=1)
original_text_score = normalized_df.loc['original_text', 'Composite Score']

# differences from the original text
normalized_df['Difference'] = (normalized_df['Composite Score'] - original_text_score).abs()

# ranking based on similarity to the original text
ranking_df = normalized_df[['Composite Score', 'Difference']].sort_values(by='Difference')
ranking_df['Rank'] = ranking_df['Difference'].rank()

print("Ranking of models based on similarity to the original text:")
print(ranking_df)

ranking_df.to_csv("model_ranking_based_on_similarity_to_original_text_LAZY JACK.csv")

# rankings for each metric independently
metric_rankings = {}

for metric in pivot_df.columns:
    metric_rankings[metric] = normalized_df[[metric]].sort_values(by=metric, ascending=False if metric_directions[metric] == "higher" else True)
    metric_rankings[metric]['Rank'] = metric_rankings[metric][metric].rank(ascending=False if metric_directions[metric] == "higher" else True)

    print(f"\nRanking based on {metric}:")
    print(metric_rankings[metric])

    # Save each metric ranking to a CSV file
    metric_rankings[metric].to_csv(f"model_ranking_based_on_{metric}_LAZY JACK.csv")
```

```python
import plotly.express as px


df = pd.read_csv("/content/LAZY JACK_readability_score.csv")

metric_groups = {
    "Count Metrics": {
        "metrics": ["Lexicon Count", "Sentence Count"],
        "direction": "higher"
    },
    "Ease of Reading Metrics": {
        "metrics": ["Flesch Reading Ease"],
        "direction": "higher"
    },
    "Grade Level Metrics": {
        "metrics": [
            "Flesch-Kincaid Grade Level", "Gunning Fog", "SMOG Index",
            "Automated Readability Index", "Coleman-Liau Index",
            "Linsear Write Formula", "Dale-Chall Readability Score"
        ],
        "direction": "lower"
    },
    "Time Metrics": {
        "metrics": ["Reading Time (seconds)"],
        "direction": "lower"
    },
    "Syllable Count Metrics": {
        "metrics": ["Syllable Count"],
        "direction": "lower"
    }
}


pivot_df = df.pivot(index='model', columns='metric', values='value')
pivot_df = pivot_df.drop(columns=['Text Standard'], errors='ignore')

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

group_composite_scores = {}
for group_name, group_info in metric_groups.items():
    group_metrics = group_info["metrics"]
    direction = group_info["direction"]

    group_metrics = [metric for metric in group_metrics if metric in normalized_df.columns]
    group_df = normalized_df[group_metrics]

    if direction == "lower":
        group_df = 1 - group_df

    group_composite_scores[group_name] = group_df.mean(axis=1)


composite_scores_df = pd.DataFrame(group_composite_scores)
# the overall composite score as the mean of group composite scores
composite_scores_df["Composite Score"] = composite_scores_df.mean(axis=1)

# the difference from the original text's composite score
original_text_score = composite_scores_df.loc['original_text', 'Composite Score']
composite_scores_df['Difference'] = (composite_scores_df['Composite Score'] - original_text_score).abs()

composite_scores_df['Rank'] = composite_scores_df['Difference'].rank()
composite_scores_df = composite_scores_df.sort_values(by='Rank')
composite_scores_df = pd.concat([composite_scores_df.loc[['original_text']], composite_scores_df.drop(['original_text'])])

print(composite_scores_df)
composite_scores_df.to_csv("grouped_composite_scores_with_difference_and_rank.csv")

fig = px.bar(
    composite_scores_df,
    x=composite_scores_df.index,
    y='Difference',
    title="Comparison of Models by Similarity to the Original 'LAZY JACK' Text",
    labels={'Difference': 'Composite Score Difference'},
    hover_data={'Rank': True},
    text='Rank',
    color='Rank',
    color_continuous_scale=px.colors.sequential.Viridis
)

fig.update_layout(
    xaxis_title='Model',
    yaxis_title='Composite Score Difference',
    xaxis_tickangle=45,
    showlegend=False,
    plot_bgcolor='white'
)

fig.update_layout(title_font_size=24, title_x=0.5)
fig.show()
```
