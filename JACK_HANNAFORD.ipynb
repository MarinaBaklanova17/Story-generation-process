```python
import os
from unify import Unify
import pandas as pd
import textstat

os.environ["UNIFY_KEY"] = "zgZRC0qqOC89cjdiDq403mtdgp-n8ULnQjK6JMiS2YE="

def generate_response(model_name, prompt, temperature=0.7):
    unify = Unify(model_name)
    response = unify.generate(
        prompt,
        temperature=temperature
    )
    return response

models = [
    "claude-3-haiku@anthropic",
    "claude-3-opus@anthropic",
    "claude-3.5-sonnet@anthropic",
    "gemini-1.5-flash@vertex-ai",
    "gemini-1.5-pro@vertex-ai",
    "gemma-2-9b-it@fireworks-ai",
    "gemma-2b-it@together-ai",
    "gpt-3.5-turbo@openai",
    "gpt-4@openai",
    "gpt-4-turbo@openai",
    "gpt-4o@openai",
    "llama-3-70b-chat@fireworks-ai",
    "llama-3-8b-chat@fireworks-ai",
    "mistral-7b-instruct-v0.3@together-ai",
    "mistral-large@aws-bedrock",
    "mistral-small@mistral-ai",
    "mixtral-8x22b-instruct-v0.1@deepinfra",
    "mixtral-8x7b-instruct-v0.1@aws-bedrock",
    "qwen-2-72b-instruct@deepinfra"
]

prompt = "You are a fairy tale author. Write a fairy tale about JACK HANNAFORD"


original_text = """
There was an old soldier who had been long in the wars—so long, that he was quite out-at-elbows, and he did not know where to go to find a living. So he walked up moors, down glens, till at last he came to a farm, from which the good man had gone away to market. The wife of the farmer was a very foolish woman, who had been a widow when he married her; the farmer was foolish enough, too, and it is hard to say which of the two was the more foolish. When you've heard my tale you may decide.
Now before the farmer goes to market says he to his wife: “Here is ten pounds all in gold, take care of it till I come home.” If the man had not been a fool he would never have given the money to his wife to keep. Well, off he went in his cart to market, and the wife said to herself: “I will keep the ten pounds quite safe from thieves;” so she tied it up in a rag, and she put the rag up the parlour chimney.
“There,” said she, “no thieves will ever find it now, that is quite sure.”
Jack Hannaford, the old soldier, came and rapped at the door.
“Who is there?” asked the wife.
“Jack Hannaford.”
“Where do you come from?”
“Paradise.”
“Lord a' mercy! and maybe you've seen my old man there,” alluding to her former husband.
“Yes, I have.”
“And how was he a-doing?” asked the goody.
“But middling; he cobbles old shoes, and he has nothing but cabbage for victuals.”
“Deary me!” exclaimed the woman. “Didn't he send a message to me?”
“Yes, he did,” replied Jack Hannaford. “He said that he was out of leather, and his pockets were empty, so you were to send him a few shillings to buy a fresh stock of leather.”
“He shall have them, bless his poor soul!” And away went the wife to the parlour chimney, and she pulled the rag with the ten pounds in it from the chimney, and she gave the whole sum to the soldier, telling him that her old man was to use as much as he wanted, and to send back the rest.
It was not long that Jack waited after receiving the money; he went off as fast as he could walk.
Presently the farmer came home and asked for his money. The wife told him that she had sent it by a soldier to her former husband in Paradise, to buy him leather for cobbling the shoes of the saints and angels of Heaven. The farmer was very angry, and he swore that he had never met with such a fool as his wife. But the wife said that her husband was a greater fool for letting her have the money.
There was no time to waste words; so the farmer mounted his horse and rode off after Jack Hannaford. The old soldier heard the horse's hoofs clattering on the road behind him, so he knew it must be the farmer pursuing him. He lay down on the ground, and shading his eyes with one hand, looked up into the sky, and pointed heavenwards with the other hand.
“What are you about there?” asked the farmer, pulling up.
“Lord save you!” exclaimed Jack: “I've seen a rare sight.”
“What was that?”
“A man going straight up into the sky, as if he were walking on a road.”
“Can you see him still?”
“Yes, I can.”
“Where?”
“Get off your horse and lie down.”
“If you will hold the horse.”
Jack did so readily.
“I cannot see him,” said the farmer.
“Shade your eyes with your hand, and you'll soon see a man flying away from you.”
Sure enough he did so, for Jack leaped on the horse, and rode away with it. The farmer walked home without his horse.
“You are a bigger fool than I am,” said the wife; “for I did only one foolish thing, and you have done two.”




"""

results_df = pd.DataFrame(columns=[
    "model", "metric", "value"
])

def calculate_metrics(text, model_name):
    metrics = {
        "Syllable Count": textstat.syllable_count(text),
        "Lexicon Count": textstat.lexicon_count(text, removepunct=True),
        "Sentence Count": textstat.sentence_count(text),
        "Flesch Reading Ease": textstat.flesch_reading_ease(text),
        "Flesch-Kincaid Grade Level": textstat.flesch_kincaid_grade(text),
        "Gunning Fog": textstat.gunning_fog(text),
        "SMOG Index": textstat.smog_index(text),
        "Automated Readability Index": textstat.automated_readability_index(text),
        "Coleman-Liau Index": textstat.coleman_liau_index(text),
        "Linsear Write Formula": textstat.linsear_write_formula(text),
        "Dale-Chall Readability Score": textstat.dale_chall_readability_score(text),
        "Text Standard": textstat.text_standard(text, float_output=False),
        "Reading Time (seconds)": textstat.reading_time(text)
    }
    return metrics

original_metrics = calculate_metrics(original_text, "original_text")
for metric, value in original_metrics.items():
    results_df = pd.concat([results_df, pd.DataFrame([{
        "model": "original_text",
        "metric": metric,
        "value": value
    }])], ignore_index=True)

for model in models:
    print(f"\nResponse from {model}:")
    try:
        response = generate_response(model, prompt)
        print(response)

        model_metrics = calculate_metrics(response, model)
        for metric, value in model_metrics.items():
            results_df = pd.concat([results_df, pd.DataFrame([{
                "model": model,
                "metric": metric,
                "value": value
            }])], ignore_index=True)

        print("\nReadability Metrics:")
        for metric, value in model_metrics.items():
            print(f"{metric}: {value}")

    except Exception as e:
        print(f"Error generating response from {model}: {e}")

results_df.to_csv("JACK HANNAFORD_readability_score.csv", index=False)
results_df.to_excel("JACK HANNAFORD_readability_score.xlsx", index=False)


print("Responses and readability metrics saved to JACK HANNAFORD_readability_score.csv, JACK HANNAFORD_readability_score.xlsx")
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("/content/JACK HANNAFORD_readability_score.csv")

metric_directions = {
    "Syllable Count": "lower",
    "Lexicon Count": "higher",
    "Sentence Count": "higher",
    "Flesch Reading Ease": "higher",
    "Flesch-Kincaid Grade Level": "lower",
    "Gunning Fog": "lower",
    "SMOG Index": "lower",
    "Automated Readability Index": "lower",
    "Coleman-Liau Index": "lower",
    "Linsear Write Formula": "lower",
    "Dale-Chall Readability Score": "lower",
    "Text Standard": "lower",
    "Reading Time (seconds)": "lower"
}

pivot_df = df.pivot(index='model', columns='metric', values='value')
if 'Text Standard' in pivot_df.columns:
    pivot_df = pivot_df.drop("Text Standard", axis=1)

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

for metric, direction in metric_directions.items():
    if direction == "lower" and metric in normalized_df.columns:
        normalized_df[metric] = 1 - normalized_df[metric]

normalized_df['Composite Score'] = normalized_df.mean(axis=1)
original_text_score = normalized_df.loc['original_text', 'Composite Score']

# differences from the original text
normalized_df['Difference'] = (normalized_df['Composite Score'] - original_text_score).abs()

# ranking based on similarity to the original text
ranking_df = normalized_df[['Composite Score', 'Difference']].sort_values(by='Difference')
ranking_df['Rank'] = ranking_df['Difference'].rank()

print("Ranking of models based on similarity to the original text:")
print(ranking_df)

ranking_df.to_csv("model_ranking_based_on_similarity_to_original_text_GOOSEY_GOOSEY_GANDER.csv")

# rankings for each metric independently
metric_rankings = {}

for metric in pivot_df.columns:
    metric_rankings[metric] = normalized_df[[metric]].sort_values(by=metric, ascending=False if metric_directions[metric] == "higher" else True)
    metric_rankings[metric]['Rank'] = metric_rankings[metric][metric].rank(ascending=False if metric_directions[metric] == "higher" else True)

    print(f"\nRanking based on {metric}:")
    print(metric_rankings[metric])

    # Save each metric ranking to a CSV file
    metric_rankings[metric].to_csv(f"model_ranking_based_on_{metric}_JACK HANNAFORD.csv")
```

```python
import plotly.express as px


df = pd.read_csv("/content/JACK HANNAFORD_readability_score.csv")

metric_groups = {
    "Count Metrics": {
        "metrics": ["Lexicon Count", "Sentence Count"],
        "direction": "higher"
    },
    "Ease of Reading Metrics": {
        "metrics": ["Flesch Reading Ease"],
        "direction": "higher"
    },
    "Grade Level Metrics": {
        "metrics": [
            "Flesch-Kincaid Grade Level", "Gunning Fog", "SMOG Index",
            "Automated Readability Index", "Coleman-Liau Index",
            "Linsear Write Formula", "Dale-Chall Readability Score"
        ],
        "direction": "lower"
    },
    "Time Metrics": {
        "metrics": ["Reading Time (seconds)"],
        "direction": "lower"
    },
    "Syllable Count Metrics": {
        "metrics": ["Syllable Count"],
        "direction": "lower"
    }
}


pivot_df = df.pivot(index='model', columns='metric', values='value')
pivot_df = pivot_df.drop(columns=['Text Standard'], errors='ignore')

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

group_composite_scores = {}
for group_name, group_info in metric_groups.items():
    group_metrics = group_info["metrics"]
    direction = group_info["direction"]

    group_metrics = [metric for metric in group_metrics if metric in normalized_df.columns]
    group_df = normalized_df[group_metrics]

    if direction == "lower":
        group_df = 1 - group_df

    group_composite_scores[group_name] = group_df.mean(axis=1)


composite_scores_df = pd.DataFrame(group_composite_scores)
# the overall composite score as the mean of group composite scores
composite_scores_df["Composite Score"] = composite_scores_df.mean(axis=1)

# the difference from the original text's composite score
original_text_score = composite_scores_df.loc['original_text', 'Composite Score']
composite_scores_df['Difference'] = (composite_scores_df['Composite Score'] - original_text_score).abs()

composite_scores_df['Rank'] = composite_scores_df['Difference'].rank()
composite_scores_df = composite_scores_df.sort_values(by='Rank')
composite_scores_df = pd.concat([composite_scores_df.loc[['original_text']], composite_scores_df.drop(['original_text'])])

print(composite_scores_df)
composite_scores_df.to_csv("grouped_composite_scores_with_difference_and_rank.csv")

fig = px.bar(
    composite_scores_df,
    x=composite_scores_df.index,
    y='Difference',
    title="Comparison of Models by Similarity to the Original 'JACK HANNAFORD' Text",
    labels={'Difference': 'Composite Score Difference'},
    hover_data={'Rank': True},
    text='Rank',
    color='Rank',
    color_continuous_scale=px.colors.sequential.Viridis
)

fig.update_layout(
    xaxis_title='Model',
    yaxis_title='Composite Score Difference',
    xaxis_tickangle=45,
    showlegend=False,
    plot_bgcolor='white'
)

fig.update_layout(title_font_size=24, title_x=0.5)
fig.show()
```
