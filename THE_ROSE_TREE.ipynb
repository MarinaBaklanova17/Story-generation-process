```python
import os
from unify import Unify
import pandas as pd
import textstat

os.environ["UNIFY_KEY"] = "zgZRC0qqOC89cjdiDq403mtdgp-n8ULnQjK6JMiS2YE="

def generate_response(model_name, prompt, temperature=0.7):
    unify = Unify(model_name)
    response = unify.generate(
        prompt,
        temperature=temperature
    )
    return response

models = [
    "claude-3-haiku@anthropic",
    "claude-3-opus@anthropic",
    "claude-3.5-sonnet@anthropic",
    "gemini-1.5-flash@vertex-ai",
    "gemini-1.5-pro@vertex-ai",
    "gemma-2-9b-it@fireworks-ai",
    "gemma-2b-it@together-ai",
    "gpt-3.5-turbo@openai",
    "gpt-4@openai",
    "gpt-4-turbo@openai",
    "gpt-4o@openai",
    "llama-3-70b-chat@fireworks-ai",
    "llama-3-8b-chat@fireworks-ai",
    "mistral-7b-instruct-v0.3@together-ai",
    "mistral-large@aws-bedrock",
    "mistral-small@mistral-ai",
    "mixtral-8x22b-instruct-v0.1@deepinfra",
    "mixtral-8x7b-instruct-v0.1@aws-bedrock",
    "qwen-2-72b-instruct@deepinfra"
]

prompt = "You are a fairy tale author. Write a fairy tale about THE ROSE-TREE"


original_text = """
There was once upon a time a good man who had two children: a girl by a first wife, and a boy by the second. The girl was as white as milk, and her lips were like cherries. Her hair was like golden silk, and it hung to the ground. Her brother loved her dearly, but her wicked stepmother hated her. “Child,” said the stepmother one day, “go to the grocer's shop and buy me a pound of candles.” She gave her the money; and the little girl went, bought the candles, and started on her return. There was a stile to cross. She put down the candles whilst she got over the stile. Up came a dog and ran off with the candles.

She went back to the grocer's, and she got a second bunch. She came to the stile, set down the candles, and proceeded to climb over. Up came the dog and ran off with the candles.

She went again to the grocer's, and she got a third bunch; and just the same happened. Then she came to her stepmother crying, for she had spent all the money and had lost three bunches of candles.

The stepmother was angry, but she pretended not to mind the loss. She said to the child: “Come, lay your head on my lap that I may comb your hair.” So the little one laid her head in the woman's lap, who proceeded to comb the yellow silken hair. And when she combed the hair fell over her knees, and rolled right down to the ground.

Then the stepmother hated her more for the beauty of her hair; so she said to her, “I cannot part your hair on my knee, fetch a billet of wood.” So she fetched it. Then said the stepmother, “I cannot part your hair with a comb, fetch me an axe.” So she fetched it.

“Now,” said the wicked woman, “lay your head down on the billet whilst I part your hair.”

Well! she laid down her little golden head without fear; and whist! down came the axe, and it was off. So the mother wiped the axe and laughed.

Then she took the heart and liver of the little girl, and she stewed them and brought them into the house for supper. The husband tasted them and shook his head. He said they tasted very strangely. She gave some to the little boy, but he would not eat. She tried to force him, but he refused, and ran out into the garden, and took up his little sister, and put her in a box, and buried the box under a rose-tree; and every day he went to the tree and wept, till his tears ran down on the box.

One day the rose-tree flowered. It was spring, and there among the flowers was a white bird; and it sang, and sang, and sang like an angel out of heaven. Away it flew, and it went to a cobbler's shop, and perched itself on a tree hard by; and thus it sang,

  “My wicked mother slew me,
  My dear father ate me,
  My little brother whom I love
  Sits below, and I sing above
    Stick, stock, stone dead.”

“Sing again that beautiful song,” asked the shoemaker. “If you will first give me those little red shoes you are making.” The cobbler gave the shoes, and the bird sang the song; then flew to a tree in front of a watchmaker's, and sang:

  “My wicked mother slew me,
  My dear father ate me,
  My little brother whom I love
  Sits below, and I sing above
    Stick, stock, stone dead.”

“Oh, the beautiful song! sing it again, sweet bird,” asked the watchmaker. “If you will give me first that gold watch and chain in your hand.” The jeweller gave the watch and chain. The bird took it in one foot, the shoes in the other, and, after having repeated the song, flew away to where three millers were picking a millstone. The bird perched on a tree and sang:

  “My wicked mother slew me,
  My dear father ate me,
  My little brother whom I love
  Sits below, and I sing above
    Stick!”

Then one of the men put down his tool and looked up from his work,

  “Stock!”

Then the second miller's man laid aside his tool and looked up,

  “Stone!”

Then the third miller's man laid down his tool and looked up,

  “Dead!”

Then all three cried out with one voice: “Oh, what a beautiful song! Sing it, sweet bird, again.” “If you will put the millstone round my neck,” said the bird. The men did what the bird wanted and away to the tree it flew with the millstone round its neck, the red shoes in one foot, and the gold watch and chain in the other. It sang the song and then flew home. It rattled the millstone against the eaves of the house, and the stepmother said: “It thunders.” Then the little boy ran out to see the thunder, and down dropped the red shoes at his feet. It rattled the millstone against the eaves of the house once more, and the stepmother said again: “It thunders.” Then the father ran out and down fell the chain about his neck.

In ran father and son, laughing and saying, “See, what fine things the thunder has brought us!” Then the bird rattled the millstone against the eaves of the house a third time; and the stepmother said: “It thunders again, perhaps the thunder has brought something for me,” and she ran out; but the moment she stepped outside the door, down fell the millstone on her head; and so she died.






"""

results_df = pd.DataFrame(columns=[
    "model", "metric", "value"
])

def calculate_metrics(text, model_name):
    metrics = {
        "Syllable Count": textstat.syllable_count(text),
        "Lexicon Count": textstat.lexicon_count(text, removepunct=True),
        "Sentence Count": textstat.sentence_count(text),
        "Flesch Reading Ease": textstat.flesch_reading_ease(text),
        "Flesch-Kincaid Grade Level": textstat.flesch_kincaid_grade(text),
        "Gunning Fog": textstat.gunning_fog(text),
        "SMOG Index": textstat.smog_index(text),
        "Automated Readability Index": textstat.automated_readability_index(text),
        "Coleman-Liau Index": textstat.coleman_liau_index(text),
        "Linsear Write Formula": textstat.linsear_write_formula(text),
        "Dale-Chall Readability Score": textstat.dale_chall_readability_score(text),
        "Text Standard": textstat.text_standard(text, float_output=False),
        "Reading Time (seconds)": textstat.reading_time(text)
    }
    return metrics

original_metrics = calculate_metrics(original_text, "original_text")
for metric, value in original_metrics.items():
    results_df = pd.concat([results_df, pd.DataFrame([{
        "model": "original_text",
        "metric": metric,
        "value": value
    }])], ignore_index=True)

for model in models:
    print(f"\nResponse from {model}:")
    try:
        response = generate_response(model, prompt)
        print(response)

        model_metrics = calculate_metrics(response, model)
        for metric, value in model_metrics.items():
            results_df = pd.concat([results_df, pd.DataFrame([{
                "model": model,
                "metric": metric,
                "value": value
            }])], ignore_index=True)

        print("\nReadability Metrics:")
        for metric, value in model_metrics.items():
            print(f"{metric}: {value}")

    except Exception as e:
        print(f"Error generating response from {model}: {e}")

results_df.to_csv("THE ROSE-TREE_readability_score.csv", index=False)
results_df.to_excel("THE ROSE-TREE_readability_score.xlsx", index=False)


print("Responses and readability metrics saved to THE ROSE-TREE_readability_score.csv, THE ROSE-TREE_readability_score.xlsx")
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("/content/THE ROSE-TREE_readability_score.csv")

metric_directions = {
    "Syllable Count": "lower",
    "Lexicon Count": "higher",
    "Sentence Count": "higher",
    "Flesch Reading Ease": "higher",
    "Flesch-Kincaid Grade Level": "lower",
    "Gunning Fog": "lower",
    "SMOG Index": "lower",
    "Automated Readability Index": "lower",
    "Coleman-Liau Index": "lower",
    "Linsear Write Formula": "lower",
    "Dale-Chall Readability Score": "lower",
    "Text Standard": "lower",
    "Reading Time (seconds)": "lower"
}

pivot_df = df.pivot(index='model', columns='metric', values='value')
if 'Text Standard' in pivot_df.columns:
    pivot_df = pivot_df.drop("Text Standard", axis=1)

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

for metric, direction in metric_directions.items():
    if direction == "lower" and metric in normalized_df.columns:
        normalized_df[metric] = 1 - normalized_df[metric]

normalized_df['Composite Score'] = normalized_df.mean(axis=1)
original_text_score = normalized_df.loc['original_text', 'Composite Score']

# differences from the original text
normalized_df['Difference'] = (normalized_df['Composite Score'] - original_text_score).abs()

# ranking based on similarity to the original text
ranking_df = normalized_df[['Composite Score', 'Difference']].sort_values(by='Difference')
ranking_df['Rank'] = ranking_df['Difference'].rank()

print("Ranking of models based on similarity to the original text:")
print(ranking_df)

ranking_df.to_csv("model_ranking_based_on_similarity_to_original_text_THE ROSE-TREE.csv")

# rankings for each metric independently
metric_rankings = {}

for metric in pivot_df.columns:
    metric_rankings[metric] = normalized_df[[metric]].sort_values(by=metric, ascending=False if metric_directions[metric] == "higher" else True)
    metric_rankings[metric]['Rank'] = metric_rankings[metric][metric].rank(ascending=False if metric_directions[metric] == "higher" else True)

    print(f"\nRanking based on {metric}:")
    print(metric_rankings[metric])

    # Save each metric ranking to a CSV file
    metric_rankings[metric].to_csv(f"model_ranking_based_on_{metric}_THE ROSE-TREE.csv")
```

```python
import plotly.express as px


df = pd.read_csv("/content/THE ROSE-TREE_readability_score.csv")

metric_groups = {
    "Count Metrics": {
        "metrics": ["Lexicon Count", "Sentence Count"],
        "direction": "higher"
    },
    "Ease of Reading Metrics": {
        "metrics": ["Flesch Reading Ease"],
        "direction": "higher"
    },
    "Grade Level Metrics": {
        "metrics": [
            "Flesch-Kincaid Grade Level", "Gunning Fog", "SMOG Index",
            "Automated Readability Index", "Coleman-Liau Index",
            "Linsear Write Formula", "Dale-Chall Readability Score"
        ],
        "direction": "lower"
    },
    "Time Metrics": {
        "metrics": ["Reading Time (seconds)"],
        "direction": "lower"
    },
    "Syllable Count Metrics": {
        "metrics": ["Syllable Count"],
        "direction": "lower"
    }
}


pivot_df = df.pivot(index='model', columns='metric', values='value')
pivot_df = pivot_df.drop(columns=['Text Standard'], errors='ignore')

scaler = MinMaxScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)

group_composite_scores = {}
for group_name, group_info in metric_groups.items():
    group_metrics = group_info["metrics"]
    direction = group_info["direction"]

    group_metrics = [metric for metric in group_metrics if metric in normalized_df.columns]
    group_df = normalized_df[group_metrics]

    if direction == "lower":
        group_df = 1 - group_df

    group_composite_scores[group_name] = group_df.mean(axis=1)


composite_scores_df = pd.DataFrame(group_composite_scores)
# the overall composite score as the mean of group composite scores
composite_scores_df["Composite Score"] = composite_scores_df.mean(axis=1)

# the difference from the original text's composite score
original_text_score = composite_scores_df.loc['original_text', 'Composite Score']
composite_scores_df['Difference'] = (composite_scores_df['Composite Score'] - original_text_score).abs()

composite_scores_df['Rank'] = composite_scores_df['Difference'].rank()
composite_scores_df = composite_scores_df.sort_values(by='Rank')
composite_scores_df = pd.concat([composite_scores_df.loc[['original_text']], composite_scores_df.drop(['original_text'])])

print(composite_scores_df)
composite_scores_df.to_csv("grouped_composite_scores_with_difference_and_rank_THE ROSE-TREE.csv")

fig = px.bar(
    composite_scores_df,
    x=composite_scores_df.index,
    y='Difference',
    title="Comparison of Models by Similarity to the Original 'THE ROSE-TREE' Text",
    labels={'Difference': 'Composite Score Difference'},
    hover_data={'Rank': True},
    text='Rank',
    color='Rank',
    color_continuous_scale=px.colors.sequential.Viridis
)

fig.update_layout(
    xaxis_title='Model',
    yaxis_title='Composite Score Difference',
    xaxis_tickangle=45,
    showlegend=False,
    plot_bgcolor='white'
)

fig.update_layout(title_font_size=24, title_x=0.5)
fig.show()
```
